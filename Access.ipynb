```python
import requests
from bs4 import BeautifulSoup

def scrape_wayfair(query):
    url = f"https://www.wayfair.com/keyword.php?keyword={query.replace(' ', '+')}"
    
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        products = soup.find_all("div", class_="ProductCardstyles__CardWrapper")
        
        for product in products:
            title = product.find("div", class_="ProductCardstyles__ProductTitle").text.strip()
            price = product.find("span", class_="ProductCardstyles__StyledPrice").text.strip()
            link = "https://www.wayfair.com" + product.find("a", class_="ProductCardstyles__CardAnchor")["href"]
            
            print(f"Title: {title}")
            print(f"Price: {price}")
            print(f"Link: {link}")
            print("\n")
    else:
        print("Error:", response.status_code)

# Get user input for the search query
user_query = input("Enter a search query: ")
scrape_wayfair(user_query)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_wayfair_product(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Extracting product name
        product_name = soup.find("h1", class_="ProductInfoHead__ProductName").text.strip()
        
        # Extracting product price
        product_price = soup.find("span", class_="Price__value").text.strip()
        
        # Extracting product image URL
        product_image = soup.find("img", class_="ProductCarousel__thumbnailImage")["src"]
        
        # Extracting product description
        product_description = soup.find("meta", attrs={"name": "description"})["content"]
        
        # Print the extracted information
        print(f"Product Name: {product_name}")
        print(f"Product Price: {product_price}")
        print(f"Product Image: {product_image}")
        print(f"Product Description: {product_description}")
    else:
        print("Error:", response.status_code)

# Wayfair product URL
wayfair_product_url = "https://www.wayfair.com/furniture/pdp/wade-logan-barrientez-85-square-arm-sofa-bed-w011204602.html"

# Scrape and display information
scrape_wayfair_product(wayfair_product_url)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_joss_and_main(query):
    url = f"https://www.jossandmain.com/s/keyword?keywords={query.replace(' ', '+')}"
    
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        products = soup.find_all("div", class_="ProductCard__productCardInner--1p1_2")
        
        for product in products:
            title = product.find("h3", class_="ProductCard__productName--2oRgj").text.strip()
            price = product.find("span", class_="ProductPrice__productPrice--1dqXC").text.strip()
            image = product.find("img", class_="ProductCard__productImage--2F8SK")["src"]
            
            print(f"Title: {title}")
            print(f"Price: {price}")
            print(f"Image: {image}")
            print("\n")
    else:
        print("Error:", response.status_code)

# Get user input for the search query
user_query = input("Enter a search query: ")
scrape_joss_and_main(user_query)
```
