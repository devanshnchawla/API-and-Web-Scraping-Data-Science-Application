# **ChatGPT API**

```python
import openai
import pandas as pd

#key hidden
model_engine = "text-davinci-002"
prompt = 'How can I help you?'

product_df = pd.DataFrame(columns=['Product', 'Price', 'Description'])

while True:
  user_input = input("> ")

  if user_input.lower() in ['exit', 'quit', 'bye']:
    print('Goodbye!')
    break

  response = openai.Completion.create(
      engine = model_engine,
      prompt = prompt + ' ' + user_input,
      max_tokens = 1024,
      n=1,
      stop=None,
      temperature = 0.7,
    )

  response_text = response.choice[0].text.strip()
  print(response_text)
```
